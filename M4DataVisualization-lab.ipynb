Plot a histogram of `ConvertedComp.`


```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

# Provide the correct file path to the SQLite database
file_path = "/Users/jangrzebielucha/Downloads/m4_survey_data.sqlite"

# Connect to SQLite database
conn = sqlite3.connect(file_path)

# Query the data from the database
query = "SELECT ConvertedComp FROM master"
df = pd.read_sql(query, conn)

# Plot the histogram
plt.hist(df['ConvertedComp'], bins=30, color='blue', edgecolor='black')
plt.title('Histogram of ConvertedComp')
plt.xlabel('ConvertedComp')
plt.ylabel('Frequency')
plt.show()

# Close the database connection
conn.close()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

# Provide the correct file path to the SQLite database
file_path = "/Users/jangrzebielucha/Downloads/m4_survey_data.sqlite"

# Connect to SQLite database
conn = sqlite3.connect(file_path)

# Query the data from the database
query = "SELECT DatabaseDesireNextYear FROM DatabaseDesireNextYear LIMIT 100"
df = pd.read_sql(query, conn)

# Close the database connection
conn.close()

# Count the occurrences of each database and select the top 5
top_databases = df['DatabaseDesireNextYear'].value_counts().head(5)

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_databases, labels=top_databases.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()
```

```python
import pandas as pd
import sqlite3

# Provide the correct file path to the SQLite database
file_path = "/Users/jangrzebielucha/Downloads/m4_survey_data.sqlite"

# Connect to SQLite database
conn = sqlite3.connect(file_path)

# Query the data from the database
query = "SELECT DatabaseDesireNextYear FROM DatabaseDesireNextYear LIMIT 100"
df = pd.read_sql(query, conn)

# Close the database connection
conn.close()

# Count the occurrences of each database and select the top 5
top_databases = df['DatabaseDesireNextYear'].value_counts().head(5)

# Check if 'MongoDB' is in the top databases
if 'MongoDB' in top_databases.index:
    # Calculate the percentage of MongoDB
    mongodb_percentage = (top_databases['MongoDB'] / top_databases.sum()) * 100
    print(f"The percentage of MongoDB in the top 5 databases is: {mongodb_percentage:.2f}%")
else:
    print("MongoDB is not in the top 5 databases.")
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns

# Provide the correct file path to the SQLite database
file_path = "/Users/jangrzebielucha/Downloads/m4_survey_data.sqlite"

# Connect to SQLite database
conn = sqlite3.connect(file_path)

# Query the data from the database
query = "SELECT Age, WorkWeekHrs, CodeRevHrs FROM master WHERE Age BETWEEN 30 AND 35"
df = pd.read_sql(query, conn)

# Close the database connection
conn.close()

# Create a stacked chart using seaborn
plt.figure(figsize=(12, 8))
sns.barplot(x='Age', y='WorkWeekHrs', data=df, color='blue', label='WorkWeekHrs')
sns.barplot(x='Age', y='CodeRevHrs', data=df, color='orange', label='CodeRevHrs')

# Customize the plot
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.xlabel('Age Group')
plt.ylabel('Median Hours')
plt.legend()
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns

# Provide the correct file path to the SQLite database
file_path = "/Users/jangrzebielucha/Downloads/m4_survey_data.sqlite"

# Connect to SQLite database
conn = sqlite3.connect(file_path)

# Query the data from the database
query = "SELECT Age, ConvertedComp FROM master WHERE Age BETWEEN 45 AND 60"
df = pd.read_sql(query, conn)

# Close the database connection
conn.close()

# Calculate median values for 'ConvertedComp'
median_values = df.groupby('Age')['ConvertedComp'].median().reset_index()

# Plot the median ConvertedComp for each age
plt.figure(figsize=(12, 8))
sns.lineplot(x='Age', y='ConvertedComp', data=median_values, marker='o', color='blue')

# Customize the plot
plt.title('Median ConvertedComp for Ages 45-60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns

# Provide the correct file path to the SQLite database
file_path = "/Users/jangrzebielucha/Downloads/m4_survey_data.sqlite"

# Connect to SQLite database
conn = sqlite3.connect(file_path)

# Query the data from the database
query = "SELECT MainBranch, COUNT(*) AS Count FROM master GROUP BY MainBranch"
df = pd.read_sql(query, conn)

# Close the database connection
conn.close()

# Plot the horizontal bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='Count', y='MainBranch', data=df, color='skyblue')

# Customize the plot
plt.title('Main Branch Distribution')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.grid(axis='x')

# Show the plot
plt.show()

```
